```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os

count = 0
for root, folders, filenames in os.walk('/kaggle/input'):
   print(root, folders)

filepath = '/kaggle/input/myfirstproject/'
data_filter_sort = pd.read_csv(filepath+'/pandas.csv')

# 创建DataFrame
df_filter_sort = pd.DataFrame(data_filter_sort)
duplicates = df_filter_sort.duplicated()
print(duplicates)

# 使用比较运算符筛选数据: 筛选年龄大于30的记录
df_age_gt_30 = df_filter_sort[df_filter_sort['Age'] > 30]

# 使用逻辑运算符结合多个条件进行筛选: 筛选年龄大于30且城市为"London"的记录
# 注意，这里修正了城市字段的大小写和变量名的语法
df_age_gt_30_london = df_filter_sort[(df_filter_sort['Age'] > 30) & 
                                     (df_filter_sort['Dept'].str.contains('M'))]

# 使用 .isin()方法筛选特定值: 筛选城市为"New York"或"Madrid"的记录
# 注意，这里修正了城市字段的大小写
df_dept_specific = df_filter_sort[df_filter_sort['Dept'].isin(['HR', 'Sales'])]

# 显示筛选结果
print("年龄大于30的记录:\n", df_age_gt_30)
print("\n年龄大于30且部门包含'M'的记录:\n", df_age_gt_30_london)
print("\n部门为'HR'或'Sales'的记录:\n", df_dept_specific)

#结果分别写入新的csv文件
df_age_gt_30.to_csv('/kaggle/working/df_age_gt_30.csv',index=False)
df_age_gt_30_london.to_csv('/kaggle/working/df_age_gt_30_london.csv',index=False)
df_dept_specific.to_csv('/kaggle/working/df_dept_specific.csv',index=False)

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```
